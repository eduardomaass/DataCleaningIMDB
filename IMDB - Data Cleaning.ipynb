# Importar as bibliotecas

```python
import pandas as pd
import datetime as dt
```

# Abrir o arquivo

```python
db = pd.read_csv('messy_IMDB_dataset.csv', sep = ';')

db
```

# Visão geral do arquivo

```python
db.info()
```

# Apagar coluna vazia

```python
db = db.drop('Unnamed: 8', 1)
db
```

# Apagar linhas vazias

```python
db.dropna(subset = ['IMBD title ID'], inplace = True)
db
```

# Limpar o nome das colunas

```python
db = db.rename(columns={'IMBD title ID': 'IMBD Title ID',
                        'Original titl�': 'Original Title', 
                        'Release year': 'Release Year',
                        'Genr�': 'Genre',
                        ' Votes': 'Votes'})

db
```

# Visão geral do arquivo

```python
db.describe()
```

# Limpar o ID

```python
db['IMBD Title ID'] = db['IMBD Title ID'].replace({'tt':''}, regex=True)

db
```

# Limpar Original Title

```python
db['Original Title'] = db['Original Title'].str.strip()
db['Original Title']
```

# Ajustar Release Year

```python
db['Release Year'].unique()
```

```python
db['temp'] = pd.to_datetime(db['Release Year'], errors='coerce')
db[db['temp'].isnull()]
```

```python
db['Release Year'] = db['Release Year'].replace({'The 6th of marzo, year 1951':'1951'}, regex=True)
db['Release Year'] = db['Release Year'].replace({'1984-02-34':'1984'}, regex=True)
db['Release Year'] = db['Release Year'].replace({'1976-13-24':'1976'}, regex=True)
```

```python
db['temp'] = pd.to_datetime(db['Release Year'], errors='coerce')
db[db['temp'].isnull()]
```

```python
db['Release Year'] = pd.to_datetime(db['Release Year'], errors='coerce')
db['Release Year'] = db['Release Year'].dt.year
db = db.drop('temp', 1)
db
```

# Ajustar Duration

```python
db['temp'] = pd.to_numeric(db['Duration'], errors='coerce')
db[db['temp'].isnull()]
```

```python
db = db.drop('temp', 1)

db['Duration'] = db['Duration'].replace({'178c':'178'}, regex=True)
db['Duration'] = pd.to_numeric(db['Duration'], errors='coerce')
db['Duration'].fillna(0, inplace = True)
db[db['Duration'].isnull()]
```

# Ajustar Country

```python
db['Country'].unique()
```

```python
db = db.replace({'Country':{'US':'USA', 'US.':'USA',
                           'New Zesland':'New Zealand',
                           'Italy1':'Italy'}})

db['Country'].unique()
```

# Ajustar Content Rating

```python
db['Content Rating'].unique()
```

```python
db['Content Rating'].fillna('Not informed', inplace = True)
db = db.replace({'Content Rating':{'Approved':'Not informed', 
                           'Not Rated':'Not informed',
                           'Unrated': 'Not informed'}})

db['Content Rating'].unique()
```

# Ajustar Income

```python
db['Income'] = db['Income'].str[2:]
db['temp'] = pd.to_numeric(db['Income'], errors='coerce')
db[db['temp'].isnull()]
```

```python
db['Income'] = db['Income'].replace({'o':'0'}, regex=True)
db['Income'] = db['Income'].replace({',':''}, regex=True)
db['temp'] = pd.to_numeric(db['Income'], errors='coerce')
db[db['temp'].isnull()]
```

```python
db['Income'] = pd.to_numeric(db['Income'], errors='coerce')
db = db.drop('temp', 1)
db
```

# Ajustar Score

```python
db['temp'] = pd.to_numeric(db['Score'], downcast='float', errors='coerce')
db[db['temp'].isnull()]
```

```python
db['Score'] = db['Score'].replace({'e-0':''}, regex=True)
db['Score'] = db['Score'].replace({',.':'.'}, regex=True)
db['Score'] = db['Score'].replace({'f':''}, regex=True)
db['Score'] = db['Score'].replace({':':''}, regex=True)
db['Score'] = db['Score'].replace({'\++':''}, regex=True)
db['Score'] = db['Score'].replace({'8..8':'8.8'}, regex=True)
db['Score'] = db['Score'].replace({'8.7.':'8.7'}, regex=True)

db['temp'] = pd.to_numeric(db['Score'], downcast='float', errors='coerce')
db[db['temp'].isnull()]
```

```python
db['Score'] = pd.to_numeric(db['Score'], downcast='float', errors='coerce')
db = db.drop('temp', 1)
db
```

# Exportar arquivo

```python
db.to_csv('clean_IMDB_dataset.csv', sep = ';')
```
